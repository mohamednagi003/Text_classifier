```python
chars  = string.ascii_letters+" "
```

# cleaning
<ol>
    <li>
        turning text into small letters
    </li>
    <li>
        removing stop words 
    </li>  
    <li>
        removing non-english char
    </li>
</ol>        


```python
def cleaning(text):
    text = text.lower()
    text= re.sub(r"\b(im)\s","",text)
    text= re.sub(r"\b(all)","",text)
    text_split = text.split(" ")
    clean_text=[]
    for word in text_split:
        if word not in st:
            clean_text.append(word)
    text=" ".join(clean_text)
    for char in text:
        if char not in chars:
            text=text.replace(char,"")
    return text                
    
```

```python
df["clean"]=df["text"].apply(cleaning)
```

```python
df.head(20)
```

# applying vectorizer so the column text could be converted to sparse matrix

```python
vector = TfidfVectorizer()
```

```python
X = vector.fit_transform(df['clean'])
```

```python
df.drop(columns=['text'],inplace=True)
```

# applying the Random forest algo

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

```python
rand = RandomForestClassifier()
```

```python
rand.fit(X_train,y_train)
```

```python
predict = rand.predict(X_test)
```

```python
print(classification_report(y_test,predict))
```

```python
accuracy_score(y_test,predict)
```
